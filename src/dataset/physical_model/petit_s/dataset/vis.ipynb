```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

from src.dataset.physical_model.petit_s.utils.petits_configs import get_config, lower_config
```

```python
idx = 35667
config = lower_config(get_config())
os.chdir(config['proj']['cwd'])
data_path = Path(config["proj"]["data_path"]) / "petits" / "priors"
wl0_path = data_path / "9um" / f"{idx}.npy"
wl1_path = data_path / "11um" / f"{idx}.npy"
```

```python
wl0 = np.load(wl0_path)
wl1 = np.load(wl1_path)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(wl0, cmap='gray')
plt.title(f'9μm (Index: {idx})')
plt.colorbar(label='Intensity')
plt.subplot(1, 2, 2)
plt.imshow(wl1, cmap='gray')
plt.title(f'11μm (Index: {idx})')
plt.colorbar(label='gray')
plt.tight_layout()
plt.show()
# %%
```
